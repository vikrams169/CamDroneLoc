```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import copy
import time
```

```python
def imshow(img):
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img1)
```

```python
img = cv2.imread('map_sq.png')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
kp = sift.detect(gray,None)
img_kp=cv2.drawKeypoints(copy.deepcopy(img),kp,copy.deepcopy(img))

imshow(img_kp)
```

```python
img_subsection = img_kp[0:100,0:100]

imshow(img_subsection)
```

```python
img2 = img[0:100,0:100]

st = time.time()
gray2 = gray[0:100,0:100]
kp = sift.detect(gray2,None)
et = time.time()

img2=cv2.drawKeypoints(img2,kp,img2)

elapsed_time = et - st
print('Execution time:', elapsed_time, 'seconds')

imshow(img2)
```
